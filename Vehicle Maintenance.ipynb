```python
import numpy as np
import math
import datetime
from datetime import date
Odometer_Readings = np.loadtxt('Odometer.txt')

slopes = np.array([])                                          #Daily Miles Driven
for i in range(0,len(Odometer_Readings)-1):
    slopes = np.append(slopes,Odometer_Readings[i+1]-Odometer_Readings[i])

cummulative = np.array([])                                     #Cummulative Miles Driven
cummulative = np.append(cummulative,slopes[0])
for i in range(0,len(slopes)-1):
    cummulative = np.append(cummulative,cummulative[i] + slopes[i+1])


miles_per_day = (cummulative[-1]-cummulative[-31])/(len(cummulative[-31:-1]))

odometer = int(Odometer_Readings[-1])


class Vehicle:
    def __init__(self,odometer):
        
        self.odometer = odometer
        self.oil_interval = 7500
        self.tire_rotation_interval = 5000
        self.coolant_interval = 150000
        self.dif_fluid_interval = 45000
        self.transmission_fluid_interval = 100000
        self.air_filter_interval = 10000
        self.cabin_air_filter_interval = 10000
        self.spark_plugs_interval = 100000        
        
        
    def calculate(self,product,interval,lastchange):

        if lastchange > self.odometer:
            print('Error: '+ product+' last service miles cannot exceed Odometer miles, check inputs.')
            return
        else:
            self.miles_to_service = interval - (self.odometer-lastchange)
        
        if self.miles_to_service < 0:
            print(product+' Service Overdue by: '+str(self.odometer - lastchange - interval)+ ' miles!')
            print("SERVICE IMMEDIATELY")
        else:
            self.next_change = self.odometer + self.miles_to_service
            
            print('Miles until '+product+' Service :' +str(self.miles_to_service))
            print('Change at '+str(self.next_change)+ ' miles.')
            
          #-------------------------------------------------------------- Calucate Date of Service based on Miles/day avg
    
            days_to_service = math.trunc((self.miles_to_service / miles_per_day))
            today = date.today()
            service_date = today + datetime.timedelta(days=days_to_service)
            service_date = service_date.strftime("%m/%d/%y")
            print('Service in '+str(days_to_service)+ ' days on: ' + str(service_date))
            
          
        return ''
         
        
        
```

```python
truck = Vehicle(odometer)

print('Driving '+str(round(miles_per_day,1))+ ' miles per day on average.')
print('Odometer: '+str(odometer))
print('------------------------------------------')
print()

print(truck.calculate('Oil & Filter',truck.oil_interval,lastchange = 154330))

print(truck.calculate('Tire Rotation',truck.tire_rotation_interval, lastchange= 152527))

print(truck.calculate('Air Filter',truck.air_filter_interval, lastchange = 146201))

print(truck.calculate('Cabin Air Filter',truck.cabin_air_filter_interval, lastchange = 146201))

print(truck.calculate('Differential Fluid', truck.dif_fluid_interval, 126985))

print(truck.calculate('Transmission Fluid', truck.transmission_fluid_interval, 0))

print(truck.calculate('Spark Plugs', truck.spark_plugs_interval,100000))

print(truck.calculate('Coolant', truck.coolant_interval, 0))

```
